```python
import numpy as np 
import matplotlib.pyplot as plt
import glob
import cv2
import os
import seaborn as sns
import pandas as pd
```

```python
print(os.listdir("C:/Users/MLME/"))
```

```python
#Capture images and labels into arrays.
#Start by creating empty lists.
train_images = []
train_labels = [] 
for directory_path in glob.glob("C:/Users/MLME/train/*"):
    label = directory_path.split("\\")[-1]
    print(label)
    for img_path in glob.glob(os.path.join(directory_path, "*.png")):
        print(img_path)
        img = cv2.imread(img_path, cv2.IMREAD_COLOR) #Reading color images
        img = cv2.resize(img, (224, 224))#Resize images
        #img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR) #Optional step. Change BGR to RGB
        train_images.append(img)
        train_labels.append(label)
```

```python
train_images = np.array(train_images)
train_labels = np.array(train_labels)
```

```python
#Do exactly the same for test/validation images
# test
test_images = []
test_labels = [] 
for directory_path in glob.glob("C:/Users/MLME/test/*"):
    column_label = directory_path.split("\\")[-1]
    for img_path in glob.glob(os.path.join(directory_path, "*.png")):
        print(img_path)
        img = cv2.imread(img_path, cv2.IMREAD_COLOR)
        img = cv2.resize(img, (224, 224))
        #img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR) #Optional
        test_images.append(img)
        test_labels.append(column_label)
```

```python
test_images = np.array(test_images)
test_labels = np.array(test_labels)
```

```python
#Encode labels from text (folder names) to integers.
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(test_labels)
test_labels_encoded = le.transform(test_labels)
le.fit(train_labels)
train_labels_encoded = le.transform(train_labels)

#Split data into test and train datasets (already split but assigning to meaningful convention)
#If you only have one dataset then split here
x_train, y_train, x_test, y_test = train_images, train_labels_encoded, test_images, test_labels_encoded

# Normalize pixel values to between 0 and 1
x_train, x_test = x_train / 255.0, x_test / 255.0
```

```python
import tensorflow as tf 
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.image import ImageDataGenerator

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, Flatten
from tensorflow.keras import Model
```

```python
vgg_model = tf.keras.applications.vgg16.VGG16()
print(type(vgg_model))
```

```python
model = keras.models.Sequential()
for layer in vgg_model.layers[0:-1]:
    model.add(layer)
for layer in model.layers:
    layer.trainable = False
#setting the layers until prediction as non trainable 
```

```python
model.add(layers.Dense(2))
#adding last layer to ensure two outputs- parameter number has dropped significantly
```

```python
# loss and optimizer
loss=tf.keras.losses.CategoricalCrossentropy(from_logits=True)
optim = keras.optimizers.Adam(learning_rate = 0.001)
metrics = [keras.metrics.BinaryAccuracy()]

model.compile(optimizer= optim, loss = loss, metrics = metrics)
```
